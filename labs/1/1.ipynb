```python
def load(path):
    with open(path) as f:
        lines = f.readlines()
        return lines
```

```python
loaded_data = load(path)
```

```python
def preprocess(data):
    res = []
    for line in data[1:]:
        res.append(line.replace("\n", ""))
    return "".join(res)
```

```python
data = preprocess(loaded_data)
```

```python
#Napište funkci, která libovolný DNA řetězec převede na komplementární tvar.
# T <-> U
def complement(seq):
    if "T" in seq:
        #to RNA
        return seq.replace("T", "U")
    else:
        #to DNA
        return seq.replace("U", "T")
```

```python
START_CODONE = ['ATG']
END_CODONE = ['TAA', 'TAG', 'TGA']
```

```python
#Napište funkci, která se pokusí nalézt všechny geny v genomu, tj. sekvence začínající start kodonem a končící end codonem.
#19
def find_genes(data, start_codones, end_codones):
    genes = []
    size = len(data)
    
    current_position = 0
    while(current_position < size):
        triplet_end = current_position + 3
        current_triplet = data[current_position:triplet_end]
        
        #check if start codone
        is_start_codone = current_triplet in start_codones
        
        if is_start_codone:
            #start to fine gene
            for current_position_2 in range(triplet_end, size, 3):
                triplet_end_2 = current_position_2 + 3 
                current_triplet_2 = data[current_position_2:triplet_end_2]                
                is_end_codone = current_triplet_2  in end_codones
                
                if is_end_codone:
                    #print('end codone', current_position, triplet_end_2)
                    genes.append(
                        {
                            "start": current_position,
                            "end": triplet_end_2,
                            "gene": data[current_position: triplet_end_2]
                        }
                    )
                    current_position = triplet_end_2
                    break
        current_position += 1
                
    print(f'Number of genes -> {len(genes)}')
    print(genes)
    return genes
```

```python
TEST = False
RUN_DATA = data[0:1000] if TEST else data
genes = find_genes(RUN_DATA, START_CODONE, END_CODONE)
```
